#### 1.2.  Loading the data
Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

Don't be afraid to use StackOverflow for help  with this.

```python
# DataFrame Transposed
properties = np.transpose(properties)
# Reset Index for ease of further mutations and set column labels to dates
properties = properties.reset_index()
properties.columns = properties.iloc[0]
properties = properties.drop(index=0)
properties.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# Named Borough and ID columns
properties = properties.rename(columns={"Unnamed: 0": "Borough"})
properties.columns.values[1] = 'ID'
properties.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# Apply pd.melt to 'tidy' data and make date and price values more accessible keeping Borough and ID columns in place
properties = pd.melt(properties,id_vars=['Borough','ID'],value_vars=properties.columns,var_name='date',value_name='avg_price')
```

```python
# Sort by date for readability and accessibility purposes
properties = properties.sort_values(['ID','date'],ignore_index=True)
properties.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# Check dtypes
# Essential to ensure that year valuable accessible from the date column
for price in properties['avg_price']:
    float(price)
properties.dtypes
# Confirms that date column holds dt objects
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
# Some index values references that were not London Boroughs
# Creating a list of unique index values of the Borough column allowed me to analyze and subset the list to include only the list of boroughs
locs = properties['Borough'].unique()
locs = locs[1:33]
# Subset the rows to only include rows with an index matching an actual borough
# Set index to facilitate accessing data for specific Boroughs
# Dropped ID column becausue Borough and ID columns hold indentical data
# Borough column more accessible and readable
properties = properties[properties['Borough'].isin(locs)]
properties = properties.set_index('Borough').drop(columns='ID')
properties
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
properties.loc['Barking & Dagenham'].plot(x='date',y='avg_price')
properties.loc['Barnet'].plot(x='date',y='avg_price')
properties.loc['Bexley'].plot(x='date',y='avg_price')
```

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column. 

To this end, you *could* apply a ***lambda function***. Your logic could work as follows:

Whether you go ahead with this is up to you. Just so long as you answer our initial brief: which boroughs of London have seen the greatest house price increase, on average, over the past two decades? 

```python
properties['year'] = properties.apply(lambda x: x['date'].year, axis=1)
properties
```

```python
# Create a column of yearly average prices, necessary to find the price ratio
# Required dropping average monthly price
# Accepted dropping monthly average because ultimately looking compare average price on year
# This action yielded an average date column, which held no useful information and was dropped
properties = properties.groupby(['Borough','year']).mean().drop(columns='date')
```

```python
properties.head()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Did not create function because average price per year calculated in last step
from collections import defaultdict
# Init default dictionary
price_ratios = defaultdict(int)
# My method to loop through index level = 0 and subset rows with index level = 1 having a value of 2018 and 1998
# Reference list of unique boroughs used to subset rows
for bor in locs:
    # Calculate ratio and store in default dict {key:val} = {bor:ratio}
    price_ratio = properties.loc[(bor,2018)] / properties.loc[(bor,1998)]
    price_ratios[bor] = price_ratio
# converted dictionary to DataFrame
ratios_df = pd.DataFrame(price_ratios)
```

```python
# Mutated the DF to retain the format of borough as index and renamed the column for readability
ratios_df = np.transpose(ratios_df)
ratios_df.columns = ['price_rat_2018_1998']
# Also named index column for readability and accessibility when merging
ratios_df.index.name = "Borough"
ratios_df.head()
```

```python
# Show the ratio of price increase for each borough
ratios_df.plot(kind='bar', y='price_rat_2018_1998', rot=90)
plt.show()
```

```python
# Show the initial average prices as of 1998
init_avg = properties.query('year == 1998').droplevel(1)
init_avg.plot(kind='bar', y='avg_price', rot=90, title='Average Price 1998 by Borough')
```

```python
# Renamed column to clarify data on df merge
init_avg.columns = ['init_avg_1998']
init_avg.head()
```

```python
# Assess any trend between the initial avg price and the ratio of increase over 2 decades
compare_trends = ratios_df.merge(init_avg, on='Borough')
compare_trends.plot(kind='scatter', x='init_avg_1998', y='price_rat_2018_1998')
```
