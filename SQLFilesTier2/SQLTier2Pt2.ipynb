```python
# import pd, sqlalchemy to query db with pandas
from sqlalchemy import create_engine
import pandas as pd
engine = create_engine("sqlite:///sqlite_db_pythonsqlite.db")
```

```python
# Q10: Produce a list of facilities with a total revenue less than 1000.
# The output of facility name and total revenue, sorted by revenue. Remember
# that there's a different cost for guests and members!
df_q10 = pd.read_sql_query("SELECT f.name AS facility,\
                            SUM(CASE\
                                WHEN b.memid = 0 THEN (b.slots * f.guestcost)\
                                ELSE (b.slots * f.membercost)\
                            END) AS total_revenue\
                            FROM Bookings AS b\
                                INNER JOIN Facilities as f\
                                ON b.facid = f.facid\
                            GROUP BY facility\
                            HAVING total_revenue < 1000\
                            ORDER BY total_revenue", engine)
df_q10
```

```python
# Q11: Produce a report of members and who recommended them 
# in alphabetic surname,firstname order
df_q11 = pd.read_sql_query("SELECT m1.firstname || ' ' || m1.surname AS member,\
                            m2.firstname || ' ' || m2.surname AS recommended_by\
                            FROM Members AS m1\
                                INNER JOIN Members AS m2\
                                ON m1.recommendedby = m2.memid", engine)
df_q11
```

```python
# Q12: Find the facilities with their usage by member, but not guests
df_q12 = pd.read_sql_query("SELECT f.name AS facility,\
                            m.firstname || ' ' || m.surname AS member,\
                            COUNT(*) AS member_total_bookings\
                            FROM\
                                (SELECT *\
                                FROM Bookings\
                                WHERE memid != 0) AS subquery\
                            INNER JOIN Facilities AS f\
                                ON subquery.facid = f.facid\
                            INNER JOIN members AS m\
                                ON subquery.memid = m.memid\
                            GROUP BY facility, member",
                            engine)
df_q12
```
